```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
from sklearn.metrics import roc_curve, roc_auc_score, classification_report, accuracy_score, confusion_matrix 
```

```python
df = pd.read_csv("train_data.csv")
len(df)
```

```python

df = df.head(10000)
```

```python
# 求年纪
df['age'] = 2022 - pd.to_datetime(df['Date.of.Birth'],format='%d/%m/%Y').dt.year
df['age']
```

```python
# 提取月份
import re
def get_month_num(text):
    pattern =r'([0-9]+)yrs\s([0-9]+)mon'
    match_obj = re.match(pattern,text)
    if match_obj is None:
        return 0
    return int(match_obj.group(1))*12+int(match_obj.group(2))
    
    
```

```python

# 将AVERAGE.ACCT.AGE , CREDIT.HISTORY.LENGTH转换成月数
df['AVERAGE.ACCT.AGE'] = df['AVERAGE.ACCT.AGE'].apply(get_month_num)
df['CREDIT.HISTORY.LENGTH'] = df['CREDIT.HISTORY.LENGTH'].apply(get_month_num)
```

```python
# 这个是我们人工凭经验选择的特征
selected_features = [
 'disbursed_amount',
 'asset_cost',
 'ltv',
 'Employment.Type',
 'MobileNo_Avl_Flag',
 'Aadhar_flag',
 'PAN_flag',
 'VoterID_flag',
 'Driving_flag',
 'Passport_flag',
 'PERFORM_CNS.SCORE',
 'PRI.NO.OF.ACCTS',
 'PRI.ACTIVE.ACCTS',
 'PRI.OVERDUE.ACCTS',
 'PRI.CURRENT.BALANCE',
 'PRI.SANCTIONED.AMOUNT',
 'PRI.DISBURSED.AMOUNT',
 'SEC.NO.OF.ACCTS',
 'SEC.ACTIVE.ACCTS',
 'SEC.OVERDUE.ACCTS',
 'SEC.CURRENT.BALANCE',
 'SEC.SANCTIONED.AMOUNT',
 'SEC.DISBURSED.AMOUNT',
 'PRIMARY.INSTAL.AMT',
 'SEC.INSTAL.AMT',
 'NEW.ACCTS.IN.LAST.SIX.MONTHS',
 'DELINQUENT.ACCTS.IN.LAST.SIX.MONTHS',
 'AVERAGE.ACCT.AGE',
 'CREDIT.HISTORY.LENGTH',
 'NO.OF_INQUIRIES',
 'age']
X = df[selected_features]
X
```

```python
# 需要的分类的标签
y = df['loan_default']
```

```python
X = pd.get_dummies(X)
X
```

```python
from sklearn.preprocessing import StandardScaler
# 对数组特征进行标准化
numerical_features = ['disbursed_amount','asset_cost','ltv','PERFORM_CNS.SCORE','PRI.NO.OF.ACCTS','PRI.ACTIVE.ACCTS','PRI.CURRENT.BALANCE',
                      'PRI.SANCTIONED.AMOUNT','PRI.DISBURSED.AMOUNT','PRIMARY.INSTAL.AMT','AVERAGE.ACCT.AGE','CREDIT.HISTORY.LENGTH','age']

st = StandardScaler()
scaled_feature_df=st.fit_transform(df[numerical_features])
scaled_feature_df = pd.DataFrame(scaled_feature_df,columns=numerical_features)
scaled_feature_df
```

```python
# 种类特征进行 onr hot 编码
category_features = [feature for feature in selected_features if feature not in numerical_features]
one_hot_df =pd.get_dummies(df[category_features])
one_hot_df
```

```python
# 2种类型的特征进行合并
X = pd.concat([scaled_feature_df,one_hot_df],axis=1)
X
```

```python
import statsmodels.api as sm
```

```python
# 构建OLS 做特征筛选
modle=sm.OLS(y,X) 
res=modle.fit() 
res.summary() #结果
```

```python
# 筛选p-value 小于0.1的特征,只用这些特征进行模型训练
good_features = ['PERFORM_CNS.SCORE','age','PRI.OVERDUE.ACCTS','SEC.OVERDUE.ACCTS',
                 'SEC.CURRENT.BALANCE','NO.OF_INQUIRIES','Employment.Type_Salaried','Employment.Type_Self employed']
```

```python
X = X[good_features]
X
```

```python
df['loan_default'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
# 数据集拆分  拆成训练集 验证集
xtrain, xtest_v, ytrain, ytest_v = train_test_split(X,y,test_size = 0.3,random_state=0,stratify=y)
```

```python
# 模型1 svm
from sklearn.pipeline import Pipeline

from sklearn.svm import LinearSVC, LinearSVR,SVC,SVR
from sklearn.metrics import precision_score,recall_score,f1_score


svm_clf = SVC(C=0.4,class_weight={1:3},probability=True)
svm_clf.fit( xtrain, ytrain )
```

```python
# 求准确率
acc = svm_clf.score(xtest_v, ytest_v)
print("svm acc is ",acc)
```

```python
f1_score_value = f1_score(svm_clf.predict(xtest_v), ytest_v)
print("svm f1_score:",f1_score_value)
```

```python
# f1_score_value = f1_score(svm_clf.predict(xtrain), ytrain)
# print("f1_score:",f1_score_value)
```

```python
y_test_hat_probs = svm_clf.predict_proba(xtest_v)[:,0]
test_auc_roc = roc_auc_score(ytest_v, y_test_hat_probs)*100
print('svm AUC: %.4f %%' % test_auc_roc)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit( xtrain, ytrain )
```

```python
# 求准确率
acc1 = knn.score(xtest_v, ytest_v)
print("knn acc is ",acc1)
```

```python
f1_score_value = f1_score(knn.predict(xtest_v), ytest_v)
print("knn f1_score:",f1_score_value)
```

```python
y_test_hat_probs = knn.predict_proba(xtest_v)[:,0]
test_auc_roc = roc_auc_score(ytest_v, y_test_hat_probs)*100
print('knn Training AUC: %.4f %%' % test_auc_roc)
```

```python
from sklearn.linear_model import LogisticRegression

scaler = StandardScaler()
lr = LogisticRegression()
lg = Pipeline([('standardize', scaler),
                    ('log_reg', lr)])
lg.fit( xtrain, ytrain )
```

```python
# 求准确率
acc2 = lg.score(xtest_v, ytest_v)
print("LogisticRegression acc is ",acc2)
```

```python
f1_score_value = f1_score(lg.predict(xtest_v), ytest_v)
print("LogisticRegression f1_score:",f1_score_value)
```

```python
y_test_hat_probs = lg.predict_proba(xtest_v)[:,1]
test_auc_roc = roc_auc_score(ytest_v, y_test_hat_probs)*100
print('LogisticRegression Training AUC: %.4f %%' % test_auc_roc)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(['svm','knn','LogisticRegression'], [acc,acc1,acc2])
plt.title("acc by model")
plt.xlabel("model name")
plt.ylabel("acc")
plt.xticks(rotation=90)
plt.show()
# 最好的模型是逻辑回归
```
